# Exponential Smoothing

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
es = ExponentialSmoothing(daily_data['Global_active_power'], trend=None, seasonal=None).fit()
pred = es.predict(start=daily_data['Global_active_power'].index[0], end=df['Global_active_power'].index[-1])
plt.plot(pred)
plt.show()
```

```python
plt.figure(figsize=(25,20))
plt.plot(daily_data['Global_active_power'])
plt.plot(pred)
plt.show()
```
